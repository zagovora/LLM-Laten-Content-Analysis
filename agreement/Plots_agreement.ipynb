```python
from nltk import agreement
from nltk.metrics.distance import masi_distance
from nltk.metrics.distance import jaccard_distance


df_tr=pd.read_csv("triplets_salamanca.csv")

data = []
for idx, row in df_tr.iterrows():
    #print (row.values)
    data.append(tuple(row.values))
```

```python
# based on https://stackoverflow.com/questions/45741934/
from nltk.metrics import interval_distance,binary_distance

jaccard_task = agreement.AnnotationTask(distance=jaccard_distance)
masi_task = agreement.AnnotationTask(distance=masi_distance)
interval_task = agreement.AnnotationTask(distance=interval_distance)
binary_task = agreement.AnnotationTask(distance=binary_distance)

tasks = [interval_task,binary_task]#, masi_task,jaccard_task]
for task in tasks:
    task.load_array(data)
    print("Statistics for dataset using {}".format(task.distance))
    print("C: {}\nI: {}\nK: {}".format(task.C, task.I, task.K))
    #print("Pi: {}".format(task.pi()))
    print("Kappa: {}".format(task.kappa()))
    print("Multi-Kappa: {}".format(task.multi_kappa()))
    print("Alpha: {}".format(task.alpha()))
```

## Human agreement - Bootstrapping

```python
import random
random.seed(1)

all_ids=[]
all_alpha=[]
all_alpha_1=[]
all_alpha_2=[]
all_alpha_3=[]
all_alpha_4=[]
for i in range(0,1000):
    ids=random.sample(range(0, 33), 11)
    all_ids.append(ids)
    df_tr_sample=df_tr[df_tr["Row"].isin(ids)]
    data_sample = []
    for idx, row in df_tr_sample.iterrows():
        data_sample.append(tuple(row.values))
    interval_task3 = agreement.AnnotationTask(distance=interval_distance)  
    interval_task3.load_array(data_sample)
    all_alpha.append(interval_task3.alpha())
    
    df_tr_sample_1=df_tr_sample[df_tr_sample["Column"].isin(range(0,25))]
    df_tr_sample_2=df_tr_sample[df_tr_sample["Column"].isin(range(25,50))]
    df_tr_sample_3=df_tr_sample[df_tr_sample["Column"].isin(range(50,75))]
    df_tr_sample_4=df_tr_sample[df_tr_sample["Column"].isin(range(75,100))]
    
    data_sample = []
    for idx, row in df_tr_sample_1.iterrows():
        data_sample.append(tuple(row.values))
    interval_task3 = agreement.AnnotationTask(distance=interval_distance)  
    interval_task3.load_array(data_sample)
    all_alpha_1.append(interval_task3.alpha())
    
    data_sample = []
    for idx, row in df_tr_sample_2.iterrows():
        data_sample.append(tuple(row.values))
    interval_task3 = agreement.AnnotationTask(distance=interval_distance)  
    interval_task3.load_array(data_sample)
    all_alpha_2.append(interval_task3.alpha())
    
    data_sample = []
    for idx, row in df_tr_sample_3.iterrows():
        data_sample.append(tuple(row.values))
    interval_task3 = agreement.AnnotationTask(distance=interval_distance)  
    interval_task3.load_array(data_sample)
    all_alpha_3.append(interval_task3.alpha())
    
    
    data_sample = []
    for idx, row in df_tr_sample_4.iterrows():
        data_sample.append(tuple(row.values))
    interval_task3 = agreement.AnnotationTask(distance=interval_distance)  
    interval_task3.load_array(data_sample)
    all_alpha_4.append(interval_task3.alpha())
```
